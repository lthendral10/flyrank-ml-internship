## Split Design

An 80/20 train-test split is used. The training data is used to fit the model, while the test data is kept separate for evaluation. This provides an honest estimate of performance because the model is evaluated on unseen data.

The same split is used when comparing the model with the Week 4 baseline.

```python
df = pd.read_csv("/content/flyrank-ml-internship/data/raw/content_refresh_anonymized.csv")

# Create a proxy target from trend_direction
df["target"] = (df["trend_direction"] == "down").astype(int)

# Feature columns
features = [
    "search_volume",
    "ctr",
    "avg_position",
    "days_since_last_update",
    "content_age_days",
    "engagement_rate",
    "scroll_rate"
]

X = df[features].fillna(0)
y = df["target"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print(X_train.shape)
print(X_test.shape)
```
